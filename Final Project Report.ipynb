<h1>Using Clustering to Find Advertising Potential for Toronto Neighborhoods</h1>

<h2>Introduction:</h2>
I want to see if there is any correlation between neighborhoods and business clusters in order to find potential advertisement opportunities.

<h2>Interested Parties:</h2>
Potential parties who would be interested in this would be small scale government groups who would seek to explore potential for some extra revenue to flow into their local neighborhood.

<h2>Data:</h2>
<p>The data used in the code comes from three sources: </p>
<p>The first is a CSV file containing the coordinates and postal code of every neighborhood in Toronto. </p>
<p>The second is data about Toronto neighborhoods that was scraped from a wikipedia table.</p>
<p>The third is data retrieved from the FourSquare API concerning various venues in each neighborhood.</p>

<h2>Methodology:</h2>
<p>First, a wikipedia page containing information about Toronto neighborhoods and postal codes was scraped using BeautifulSoup 4 to get any relevant data and converted into a Pandas DataFrame. Then, additional data from a seperate CSV file was combined to get specific latitudes and longitudes. The Toronto neighborhood location DataFrame was fed into a KMeans clustering algorithm, and then displayed on a folium map to visualize said clusters.  </p>

<p> After exploring potential neighborhood clusters, the FourSquare API was used to get information about popular venues for each neighborhood using the explore query. This information was appended to a main DataFrame, then cleaned to improve readability. This included removing unimportant columns such as ...... and ......, as well as deleting any duplicate venues that may have appeared, since the search radius of each neighborhood was set to 2 miles. Finally, the data was fed into a KMeans clustering algorithm, and similar to the Toronto neighborhoods data, displayed on a Folium map.</p>

<h2>Results:</h2>
Unfortunately, a free FourSquare account lacked the call capability to properly return all queries for all neighborhoods, so the data displayed on the map is incomplete. Therefore, we cannot come to any meaningful conclusions. However, with more calls to the API, some correlation may be found.
