<h1>Final Project Report</h1>

<h2>Introduction/Business Problem</h2>

What I will be attempting to do is to help communities in 
Toronto be able to advertise their strengths in terms of the 
venues which make their home in said communities. For example, 
communities which house many restaurants could use the fact to 
draw more visitors to their shops. Additionally, I will see if 
venue clusters match up with community clusters.

<h2>Data</h2>

There will be two different sets of data used. The first is the 
Week 3 toronto neighborhood dataframe. The second is data of venues 
gathered from the foursquare api. This data will be used to form 
clusters of both neighborhoods as well as venues, of which will 
be interpreted to solve the problem presented.
